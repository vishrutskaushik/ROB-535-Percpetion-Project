```python
# label = pd.read_csv("prediction_outputs1.csv", delimiter=",")
# label = pd.read_csv("prediction_outputs_vgg19_1.csv", delimiter=",")
# label = pd.read_csv("prediction_outputs_inception3-noweights.csv", delimiter=",")
# label.info()

label_resnet = pd.read_csv("prediction_outputs_resnet-noweights.csv", delimiter=",")
label_vgg19 = pd.read_csv("prediction_outputs_vgg19.csv", delimiter=",")
label_inceptionv3 = pd.read_csv("prediction_outputs_inceptionv3.csv", delimiter=",")
label_resnet.info()
label_vgg19.info()
label_inceptionv3.info()
```

```python
# output = ["guid/image,label\n"] + [abc[i][20:-15] + "/" + abc[i][-14:-10] + "," + str(label["0"][i]) + "\n" for i in range(len(abc))]


output = ["guid/image,label\n"]
for i in range(len(abc)):
    if label_inceptionv3["0"][i] == 0 or label_vgg19["0"][i] == 0 or label_resnet["0"][i] == 0:
        label = 0
    else:
        label = int(np.round((label_inceptionv3["0"][i] + label_vgg19["0"][i] + label_resnet["0"][i]) / 3))
    output.append(abc[i][20:-15] + "/" + abc[i][-14:-10] + "," + str(label) + "\n")

# output = ["guid/image,label\n"]
# for i in range(len(abc)):
#     if label_inceptionv3["0"][i] == 0 or label_vgg19["0"][i] == 0 or label_resnet["0"][i] == 0:
#         label = 0
    
#     if label_inceptionv3["0"][i] == label_vgg19["0"][i] and label_inceptionv3["0"][i] == label_resnet["0"][i]:
#         label = label_inceptionv3["0"][i]
#     else:
#         label = 2
#     output.append(abc[i][20:-15] + "/" + abc[i][-14:-10] + "," + str(int(label)) + "\n")
```
